```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt


x = sp.Symbol('x')
y = sp.Function('y')

# ============================================================
# 1. y' = x*y , con y(2) = 4
# ============================================================
print("1) y' = x*y , con y(2) = 4")

edo1 = sp.Eq(sp.diff(y(x), x), x * y(x))
sol1 = sp.dsolve(edo1, ics={y(2): 4})
display(sol1)

y1_sol = sp.lambdify(x, sol1.rhs, 'numpy')
x_vals = np.linspace(0, 4, 100)
y_vals = y1_sol(x_vals)

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='y = '+str(sol1.rhs))
plt.title("Solución de y' = x*y")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 2. y' = 2y , con y(0) = 1
# ============================================================
print("2) y' = 2y , con y(0) = 1")

edo2 = sp.Eq(sp.diff(y(x), x), 2*y(x))
sol2 = sp.dsolve(edo2, ics={y(0): 1})
display(sol2)

y2_sol = sp.lambdify(x, sol2.rhs, 'numpy')
x_vals = np.linspace(0, 3, 100)
y_vals = y2_sol(x_vals)

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='y = '+str(sol2.rhs))
plt.title("Solución de y' = 2y")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 3. y' = (x - y)/(x + y) , con y(1) = 4
# ============================================================
print("3) y' = (x - y)/(x + y) , con y(1) = 4")

edo3 = sp.Eq(sp.diff(y(x), x), (x - y(x)) / (x + y(x)))
sol3 = sp.dsolve(edo3, ics={y(1): 4})
display(sol3)

y3_sol = sp.lambdify(x, sol3.rhs, 'numpy')
x_vals = np.linspace(0.1, 6, 200)
y_vals = y3_sol(x_vals)

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='y = '+str(sol3.rhs))
plt.title("Solución de y' = (x - y)/(x + y)")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)
plt.show()
```
